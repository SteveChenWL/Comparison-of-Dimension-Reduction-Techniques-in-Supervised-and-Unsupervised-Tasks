### Parallel Process to get wikipedia data
Here, we will request data from wikipedia-api by parallel computing.

```python
from threading import Thread
import pandas as pd
import wikipedia
import wikipediaapi
import wptools
import warnings
#import multiprocessing as mp

warnings.filterwarnings("ignore")
wiki_wiki = wikipediaapi.Wikipedia('en')
book = pd.read_csv('book_names.csv')
books = book['name'][:50]

#print("Number of processors: ", mp.cpu_count())
```

Reference for multi-threads:https://nbviewer.jupyter.org/gist/minrk/5732094

```python
# define functions to get data from wikipedia api

def get_sections(bookname, level=0):
    p = wiki_wiki.page(bookname)
    sections = p.sections
    titles = []
    texts = []
    for s in sections:
        titles.append(s.title)
        texts.append(s.text)
    return dict(zip(titles, texts))  

def get_content(book):
    try:
        page = wikipedia.page(book)
        title.append(page.title)
        return get_sections(book)   
    except:
        pass
    
def get_wikipedia(books,book_dict = None):
    #process a number of ids, storing the results in a dict
    if book_dict is None:
        book_dict = {}
    for b in books:
        book_dict[b] = get_content(b)  
    return book_dict
    
# thread process
def threaded_process(books,nthreads):
    #process the id range in a specified number of threads
    book_dict = {}
    threads = []
    id_range = (range(len(books))) 
    # create the threads
    for i in range(nthreads):
        b = books[i::nthreads]
        t = Thread(target = get_wikipedia,args =(b,book_dict))
        threads.append(t)
    
    # start the threads
    [ t.start() for t in threads]
    # wait for the threads to finish
    [ t.join() for t in threads]
    return book_dict
```
